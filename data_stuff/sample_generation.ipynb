```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def circular(center = (0, 0), radius = 1.0, size = 100):
    
    retval = np.empty((size, 2))
    r = np.sqrt(np.random.random(size) * radius * radius)
    a = np.random.random(size) * 2 * math.pi
    
    retval[:, 0] =  center[0] + np.sin(a) * r
    retval[:, 1] =  center[1] + np.cos(a) * r
    
    return retval
```

```python
X = circular(center = (3, 4), radius = 5, size = 500)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```
